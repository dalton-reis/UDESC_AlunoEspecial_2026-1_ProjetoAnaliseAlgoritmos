Para mais detalhes dos algoritmos implementados abaixo ver [README.md](../README.md):

1. **AntesSequencia**: le `m` antes da sequencia e conta durante a leitura.
2. **DepoisSequencia**: le toda a sequencia primeiro e conta `m` depois.

As funções de leitura validam entrada inteira e inteira positiva.

```python
def ler_inteiro():
    while True:
        entrada = input()
        try:
            return int(entrada)
        except ValueError:
            print("Valor invalido! Digite um numero inteiro.")


def ler_inteiro_positivo():
    while True:
        entrada = input()
        try:
            numero = int(entrada)
            if numero > 0:
                return numero
            print("Valor invalido! Digite um numero inteiro positivo.")
        except ValueError:
            print("Valor invalido! Digite um numero inteiro positivo.")


def antes_sequencia():
    print("Informe a qtd. de valores (n) menor que 1.000.000: ", end="")
    n = ler_inteiro_positivo()

    print("Informe o valor a buscar (m): ", end="")
    m = ler_inteiro()

    contador = 0  # contador de ocorrências de m em n.

    print(f"Informe os {n} valores de (n) ____")
    for _ in range(n):
        x = ler_inteiro()
        if x == m:
            contador += 1

    print(f"O valor (m) apareceu {contador} vezes em (n).")


def depois_sequencia():
    print("Informe a qtd. de valores (n) menor que 1.000.000: ", end="")
    n = ler_inteiro_positivo()

    print(f"Informe os {n} valores de (n) ____")
    vetor_n = [0] * n
    for i in range(n):
        vetor_n[i] = ler_inteiro()

    print("Informe o valor a buscar (m): ", end="")
    m = ler_inteiro()

    contador = 0  # contador de ocorrências de m em n.
    for i in range(n):
        if vetor_n[i] == m:
            contador += 1

    print(f"O valor (m) apareceu {contador} vezes em (n).")


def main():
    antes_sequencia()
    depois_sequencia()


main()
```
