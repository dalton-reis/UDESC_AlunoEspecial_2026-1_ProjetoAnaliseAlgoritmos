```python
# Esta célula importa os módulos usados nas simulações e visualizações.
# sortedcontainers e opcional: se nao estiver instalado, execute 'pip install sortedcontainers'.
from collections import deque
from time import perf_counter
import random

try:
    from sortedcontainers import SortedList
    SORTEDCONTAINERS_OK = True
except Exception:
    SORTEDCONTAINERS_OK = False
    SortedList = None
```

```python
# list: implementação das operações.
# Complexidades (media): inserir no fim O(1) amortizado, pesquisar O(n), remover por valor O(n).
def list_insert(container, value):
    container.append(value)

def list_search(container, value):
    try:
        return container.index(value)
    except ValueError:
        return -1

def list_remove(container, value):
    try:
        container.remove(value)
        return True
    except ValueError:
        return False
```

```python
# list: simulação de teste para inserir, pesquisar e remover (foco em comportamento e tempo).
N = 20000
dados = list(range(N))
alvo = N - 1

lst = []
t0 = perf_counter()
for x in dados:
    list_insert(lst, x)
t_insert = perf_counter() - t0

t0 = perf_counter()
idx = list_search(lst, alvo)
t_search = perf_counter() - t0

t0 = perf_counter()
ok = list_remove(lst, alvo)
t_remove = perf_counter() - t0

print("[list] inserir:", f"{t_insert:.6f}s")
print("[list] pesquisar índice:", idx, f"em {t_search:.6f}s")
print("[list] remover sucesso:", ok, f"em {t_remove:.6f}s")
```

```python
# list: visualização didática passo a passo das operações.
lst_demo = [10, 20, 30]
print("Estado inicial:", lst_demo)

list_insert(lst_demo, 40)
print("Depois de inserir 40:", lst_demo)

pos = list_search(lst_demo, 30)
print("Pesquisar 30 -> índice:", pos)

list_remove(lst_demo, 20)
print("Depois de remover 20:", lst_demo)
```

```python
# deque: implementação das operações.
# Complexidades (media): inserir no fim O(1), pesquisar O(n), remover por valor O(n).
def deque_insert(container, value):
    container.append(value)

def deque_search(container, value):
    for i, item in enumerate(container):
        if item == value:
            return i
    return -1

def deque_remove(container, value):
    try:
        container.remove(value)
        return True
    except ValueError:
        return False
```

```python
# deque: simulação de teste para inserir, pesquisar e remover.
N = 20000
dados = list(range(N))
alvo = N - 1

dq = deque()
t0 = perf_counter()
for x in dados:
    deque_insert(dq, x)
t_insert = perf_counter() - t0

t0 = perf_counter()
idx = deque_search(dq, alvo)
t_search = perf_counter() - t0

t0 = perf_counter()
ok = deque_remove(dq, alvo)
t_remove = perf_counter() - t0

print("[deque] inserir:", f"{t_insert:.6f}s")
print("[deque] pesquisar índice:", idx, f"em {t_search:.6f}s")
print("[deque] remover sucesso:", ok, f"em {t_remove:.6f}s")
```

```python
# deque: visualização didática passo a passo das operações.
dq_demo = deque([10, 20, 30])
print("Estado inicial:", list(dq_demo))

deque_insert(dq_demo, 40)
print("Depois de inserir 40:", list(dq_demo))

pos = deque_search(dq_demo, 30)
print("Pesquisar 30 -> índice:", pos)

deque_remove(dq_demo, 20)
print("Depois de remover 20:", list(dq_demo))
```

```python
# sortedcontainers.SortedList: implementação das operações.
# Complexidades (media): inserir O(log n), pesquisar O(log n), remover O(log n).
def sortedlist_insert(container, value):
    container.add(value)

def sortedlist_search(container, value):
    i = container.bisect_left(value)
    if i < len(container) and container[i] == value:
        return i
    return -1

def sortedlist_remove(container, value):
    i = container.bisect_left(value)
    if i < len(container) and container[i] == value:
        container.pop(i)
        return True
    return False
```

```python
# sortedcontainers.SortedList: simulação de teste para inserir, pesquisar e remover.
if not SORTEDCONTAINERS_OK:
    print("sortedcontainers nao instalado. Execute: pip install sortedcontainers")
else:
    N = 20000
    dados = list(range(N))
    random.shuffle(dados)
    alvo = N - 1

    sl = SortedList()
    t0 = perf_counter()
    for x in dados:
        sortedlist_insert(sl, x)
    t_insert = perf_counter() - t0

    t0 = perf_counter()
    idx = sortedlist_search(sl, alvo)
    t_search = perf_counter() - t0

    t0 = perf_counter()
    ok = sortedlist_remove(sl, alvo)
    t_remove = perf_counter() - t0

    print("[SortedList] inserir:", f"{t_insert:.6f}s")
    print("[SortedList] pesquisar índice:", idx, f"em {t_search:.6f}s")
    print("[SortedList] remover sucesso:", ok, f"em {t_remove:.6f}s")
```

```python
# sortedcontainers.SortedList: visualização didática passo a passo das operações.
if not SORTEDCONTAINERS_OK:
    print("sortedcontainers nao instalado. Execute: pip install sortedcontainers")
else:
    sl_demo = SortedList([30, 10, 20])
    print("Estado inicial (ordenado):", list(sl_demo))

    sortedlist_insert(sl_demo, 40)
    print("Depois de inserir 40:", list(sl_demo))

    pos = sortedlist_search(sl_demo, 30)
    print("Pesquisar 30 -> índice:", pos)

    sortedlist_remove(sl_demo, 20)
    print("Depois de remover 20:", list(sl_demo))
```

```python
# set: implementação das operações.
# Complexidades (media): inserir O(1), pesquisar O(1), remover O(1).
def set_insert(container, value):
    container.add(value)

def set_search(container, value):
    return value in container

def set_remove(container, value):
    if value in container:
        container.remove(value)
        return True
    return False
```

```python
# set: simulação de teste para inserir, pesquisar e remover.
N = 20000
dados = list(range(N))
alvo = N - 1

st = set()
t0 = perf_counter()
for x in dados:
    set_insert(st, x)
t_insert = perf_counter() - t0

t0 = perf_counter()
encontrado = set_search(st, alvo)
t_search = perf_counter() - t0

t0 = perf_counter()
ok = set_remove(st, alvo)
t_remove = perf_counter() - t0

print("[set] inserir:", f"{t_insert:.6f}s")
print("[set] pesquisar encontrado:", encontrado, f"em {t_search:.6f}s")
print("[set] remover sucesso:", ok, f"em {t_remove:.6f}s")
```

```python
# set: visualização didàtica passo a passo das operações.
st_demo = {10, 20, 30}
print("Estado inicial:", st_demo)

set_insert(st_demo, 40)
print("Depois de inserir 40:", st_demo)

achou = set_search(st_demo, 30)
print("Pesquisar 30 ->", achou)

set_remove(st_demo, 20)
print("Depois de remover 20:", st_demo)
```

```python
# dict: implementação das operações.
# Complexidades (media): inserir O(1), pesquisar O(1), remover O(1).
def dict_insert(container, key, value):
    container[key] = value

def dict_search(container, key):
    return container.get(key, None)

def dict_remove(container, key):
    if key in container:
        del container[key]
        return True
    return False
```

```python
# dict: simulação de teste para inserir, pesquisar e remover.
N = 20000
dados = list(range(N))
chave_alvo = N - 1

mp = {}
t0 = perf_counter()
for x in dados:
    dict_insert(mp, x, x * 10)
t_insert = perf_counter() - t0

t0 = perf_counter()
valor = dict_search(mp, chave_alvo)
t_search = perf_counter() - t0

t0 = perf_counter()
ok = dict_remove(mp, chave_alvo)
t_remove = perf_counter() - t0

print("[dict] inserir:", f"{t_insert:.6f}s")
print("[dict] pesquisar valor:", valor, f"em {t_search:.6f}s")
print("[dict] remover sucesso:", ok, f"em {t_remove:.6f}s")
```

```python
# dict: visualização didática passo a passo das operações.
mp_demo = {1: 'A', 2: 'B', 3: 'C'}
print("Estado inicial:", mp_demo)

dict_insert(mp_demo, 4, 'D')
print("Depois de inserir (4, 'D'):", mp_demo)

valor = dict_search(mp_demo, 3)
print("Pesquisar chave 3 -> valor:", valor)

dict_remove(mp_demo, 2)
print("Depois de remover chave 2:", mp_demo)
```
