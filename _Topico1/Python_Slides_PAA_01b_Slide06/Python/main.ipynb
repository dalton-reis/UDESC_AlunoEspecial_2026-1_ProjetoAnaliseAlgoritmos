# Exemplo 1

Multiplicação de matrizes em Python

```python
import time

try:
    import numpy as np
except ModuleNotFoundError:
    np = None


def multiplicacao_matriz(A, B):
    if len(A[0]) != len(B):
        raise ValueError("Erro: matrizes incompatíveis")

    lins_A = len(A)
    cols_A = len(A[0])
    cols_B = len(B[0])

    C = [[0 for _ in range(cols_B)] for _ in range(lins_A)]

    for i in range(lins_A):
        for j in range(cols_B):
            for k in range(cols_A):
                C[i][j] += A[i][k] * B[k][j]
    return C


def multiplicacao_matriz_numpy(A_np, B_np):
    if np is None:
        raise ModuleNotFoundError("numpy nao esta instalado no ambiente")

    C = A_np @ B_np
    return C.tolist()


def soma_matriz(A, B):
    if len(A) != len(B) or len(A[0]) != len(B[0]):
        raise ValueError("Erro: matrizes incompatíveis para soma")

    return [
        [A[i][j] + B[i][j] for j in range(len(A[0]))]
        for i in range(len(A))
    ]


def soma_total_matriz(A):
    total = 0
    for linha in A:
        for valor in linha:
            total += valor
    return total


def criar_matriz_zeros(ordem):
    return [[0 for _ in range(ordem)] for _ in range(ordem)]


def criar_matriz_ordenada(ordem):
    matriz = []
    valor = 1
    for _ in range(ordem):
        linha = []
        for _ in range(ordem):
            linha.append(valor)
            valor += 1
        matriz.append(linha)
    return matriz


def criar_matriz_identidade(ordem):
    return [
        [1 if i == j else 0 for j in range(ordem)]
        for i in range(ordem)
    ]


def operacao_matriz_10x10_manual():
    matriz_zeros = criar_matriz_zeros(10)
    matriz_ordenada = criar_matriz_ordenada(10)
    matriz_identidade = criar_matriz_identidade(10)

    matriz_somada = soma_matriz(matriz_zeros, matriz_ordenada)
    matriz_resultante = multiplicacao_matriz(matriz_somada, matriz_identidade)
    soma_final = soma_total_matriz(matriz_resultante)

    return matriz_resultante, soma_final


def operacao_matriz_10x10_numpy():
    if np is None:
        raise ModuleNotFoundError("numpy nao esta instalado no ambiente")

    matriz_zeros = np.zeros((10, 10), dtype=int)
    matriz_ordenada = np.arange(1, 101, dtype=int).reshape(10, 10)
    matriz_identidade = np.eye(10, dtype=int)

    matriz_resultante = (matriz_zeros + matriz_ordenada) @ matriz_identidade
    soma_final = int(np.sum(matriz_resultante))

    return matriz_resultante.tolist(), soma_final


def comparar_tempos(rotulo, funcao_manual, funcao_numpy, repeticoes):
    inicio = time.perf_counter()
    for _ in range(repeticoes):
        resultado_manual = funcao_manual()
    fim = time.perf_counter()
    tempo_manual = fim - inicio

    print(f"{rotulo} manual:")
    print(resultado_manual)
    print(f"Tempo gasto: {tempo_manual:.10f} segundos")
    print()

    if np is None:
        print(f"{rotulo} com numpy: indisponivel (numpy nao instalado)")
        print()
        return

    inicio = time.perf_counter()
    for _ in range(repeticoes):
        resultado_numpy = funcao_numpy()
    fim = time.perf_counter()
    tempo_numpy = fim - inicio

    print(f"{rotulo} com numpy:")
    print(resultado_numpy)
    print(f"Tempo gasto: {tempo_numpy:.10f} segundos")
    print()

    if tempo_numpy < tempo_manual:
        diferenca_percentual = ((tempo_manual - tempo_numpy) / tempo_manual) * 100
        print(f"Numpy foi mais rapido por {diferenca_percentual:.2f}%.")
    else:
        diferenca_percentual = ((tempo_numpy - tempo_manual) / tempo_numpy) * 100
        print(f"A versao manual foi mais rapida por {diferenca_percentual:.2f}%.")
    print()


def main():
    A = [[1, 2, 3], [4, 5, 6]]
    B = [[7, 8], [9, 10], [11, 12]]

    def multiplicacao_manual_exemplo():
        return multiplicacao_matriz(A, B)

    def multiplicacao_numpy_exemplo():
        A_np = np.array(A)
        B_np = np.array(B)
        return multiplicacao_matriz_numpy(A_np, B_np)

    comparar_tempos("Multiplicacao de matrizes", multiplicacao_manual_exemplo, multiplicacao_numpy_exemplo, 1000000)
    comparar_tempos("Operacao com matriz 10x10", operacao_matriz_10x10_manual, operacao_matriz_10x10_numpy, 10000)


main()
```
