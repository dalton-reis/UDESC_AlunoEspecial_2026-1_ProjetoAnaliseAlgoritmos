# Enunciado

Use o método mestre para fornecer limites assintóticos restritos para as seguintes recorrências:

$$
\text{a) } T(n) = T(n/2) + \Theta(1)
$$

$$
\text{b) } T(n) = 4T(n/2) + \Theta(n)
$$

$$
\text{c) } T(n) = 4T(n/2) + \Theta(n^3)
$$


# Teorema Mestre

O Teorema Mestre e aplicado a recorrências da forma

$$
T(n) = aT(n/b) + f(n)
$$

comparando $f(n)$ com $n^{\log_b a}$. Os tres casos mostrados sao:

- Caso 1: se $f(n) = \Theta\left(n^{\log_b a}\right)$, então

$$
T(n) = \Theta\left(n^{\log_b a} \log n\right)
$$

- Caso 2: se $f(n) = O\left(n^{\log_b a - \varepsilon}\right)$, então

$$
T(n) = \Theta\left(n^{\log_b a}\right)
$$

- Caso 3: se $f(n) = \Omega\left(n^{\log_b a + \varepsilon}\right)$, então

$$
T(n) = \Theta(f(n))
$$

Casos especiais:

Se as condições do Caso 1, Caso 2 ou Caso 3 não forem satisfeitas, então não e possível resolver a recorrência usando o Teorema Mestre.

Exemplo:

$$
T(n) = 2T(n/2) + n \log n
$$


# Resolução -- Teorema Mestre

Células separadas para as atividades a, b e c.


Resumo das repostas

- a) $\Theta(\log n)$
- b) $\Theta(n^2)$
- c) $\Theta(n^3)$


## a) $T(n) = T(n/2) + \Theta(1)$

Aqui:

- $a = 1$
- $b = 2$
- $f(n) = \Theta(1)$

Primeiro calculamos:

$$
n^{\log_b a} = n^{\log_2 1} = n^0 = 1
$$

Logo:

$$
f(n) = \Theta(1) = \Theta(n^0) = \Theta\left(n^{\log_2 1}\right)
$$

Então a recorrência se encaixa no Caso 1 do Teorema Mestre. Pelo slide:

$$
T(n) = \Theta\left(n^{\log_b a} \log n\right)
$$

Substituindo:

$$
T(n) = \Theta\left(n^0 \log n\right) = \Theta(\log n)
$$

Resposta da atividade (a):

$$
\boxed{T(n) = \Theta(\log n)}
$$


## b) $T(n) = 4T(n/2) + \Theta(n)$

Aqui:

- $a = 4$
- $b = 2$
- $f(n) = \Theta(n)$

Calculamos:

$$
n^{\log_b a} = n^{\log_2 4} = n^2
$$

Agora comparamos:

$$
f(n) = \Theta(n)
$$

com

$$
n^{\log_2 4} = n^2
$$

Como $n$ cresce mais devagar que $n^2$, temos:

$$
\Theta(n) = O(n^{2 - \varepsilon})
$$

com, por exemplo, $\varepsilon = 1$, pois:

$$
n^{2 - 1} = n
$$

Portanto, a recorrência se encaixa no Caso 2 do Teorema Mestre. Pelo slide:

$$
T(n) = \Theta\left(n^{\log_b a}\right)
$$

Substituindo:

$$
T(n) = \Theta(n^2)
$$

Resposta da atividade (b):

$$
\boxed{T(n) = \Theta(n^2)}
$$


## c) $T(n) = 4T(n/2) + \Theta(n^3)$

Aqui:

- $a = 4$
- $b = 2$
- $f(n) = \Theta(n^3)$

Calculamos novamente:

$$
n^{\log_b a} = n^{\log_2 4} = n^2
$$

Comparando:

$$
f(n) = \Theta(n^3)
$$

com

$$
n^{\log_2 4} = n^2
$$

Como $n^3$ cresce mais rápido que $n^2$, então:

$$
\Theta(n^3) = \Omega(n^{2 + \varepsilon})
$$

com $\varepsilon = 1$, pois:

$$
n^{2 + 1} = n^3
$$

Assim, a recorrência se candidata ao Caso 3. Falta verificar a condição de regularidade do slide:

$$
a\,f(n/b) \leq c\,f(n), \quad \text{para alguma constante } c < 1
$$

Substituindo:

$$
a\,f(n/b) = 4 \cdot (n/2)^3
$$

$$
= 4 \cdot \frac{n^3}{8} = \frac{n^3}{2}
$$

Como $f(n) = n^3$, resulta:

$$
a\,f(n/b) = \frac{1}{2}f(n)
$$

Basta tomar $c = \frac{1}{2}$, e como $\frac{1}{2} < 1$, a condição e satisfeita.

Então, pelo Caso 3:

$$
T(n) = \Theta(f(n)) = \Theta(n^3)
$$

Resposta da atividade (c):

$$
\boxed{T(n) = \Theta(n^3)}
$$
