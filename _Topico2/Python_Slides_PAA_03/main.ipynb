## Exemplo: Fatorial
Slide 07

````python
def fatorial(n: int) -> int:
    if n <= 1:
        return 1
    return n * fatorial(n - 1)

print("Fatorial:", fatorial(5))
````

````python
def pesqbin(v: list[int], p: int, r: int, e: int) -> int:
    if r < p:
        return -1
    q = (p + r) // 2
    if e == v[q]:
        return q
    if e < v[q]:
        return pesqbin(v, p, q - 1, e)
    return pesqbin(v, q + 1, r, e)

main_busca_binaria()
````

# fib(n) recursiva:

- Chama fib(n-1) + fib(n-2) até chegar nos casos base.  
- É a versão mais parecida com a definição matemática.  
- Repete muito trabalho. fib(30) recalcula vários mesmos valores muitas vezes.  
- Custo: tempo exponencial, aproximadamente O(2^n).  
- Memória: O(n) pela pilha de chamadas.  
- No teste foi muito mais lenta: cerca de 0.16 s.

```text
fib (5)  
fib(4) + fib(3)  
fib (3)+fib(2) + fib(2)+fib(1)  
fib(2)+fib(1) + fib(1)+1 + fib(1)+1+ 1  
```

````python
import time

def fib(n: int) -> int:
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fib(n - 1) + fib(n - 2)

inicio = time.perf_counter()
resultado = fib(30)
fim = time.perf_counter()

print("Fibonacci:", resultado)
print(f"Tempo com recursividade: {fim - inicio:.6f} segundos")
````

# fib_iterativo(n):

- Usa um laço e só guarda os dois últimos valores (a e b).
- Não repete contas.
- Custo: tempo linear O(n).
- Memória: constante O(1).
- É a melhor versão se o objetivo é só obter o valor final.
- No teste ficou em torno de 0.000028 s.

````python
import time

def fib_iterativo(n: int) -> int:
    if n == 0:
        return 0
    if n == 1:
        return 1

    a, b = 0, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    return b

inicio = time.perf_counter()
resultado = fib_iterativo(30)
fim = time.perf_counter()

print("Fibonacci sem recursividade:", resultado)
print(f"Tempo sem recursividade: {fim - inicio:.6f} segundos")
````

# fib_vetor(n):

- Também é linear, mas guarda todos os resultados anteriores em uma lista fibs.
- Custo: tempo linear O(n).
- Memória: linear O(n).
- É útil quando você quer acessar vários Fibonacci já calculados, não só o último.
- Teste ficou quase igual à iterativa, mas usa mais memória.

````python
import time

def fib_vetor(n: int) -> int:
    if n == 0:
        return 0
    if n == 1:
        return 1

    fibs = [0] * (n + 1)
    fibs[0] = 0
    fibs[1] = 1

    for i in range(2, n + 1):
        fibs[i] = fibs[i - 1] + fibs[i - 2]

    return fibs[n]

inicio = time.perf_counter()
resultado = fib_vetor(30)
fim = time.perf_counter()

print("Fibonacci com vetor:", resultado)
print(f"Tempo com vetor: {fim - inicio:.6f} segundos")
````
